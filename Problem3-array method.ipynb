```python
def read_data(f):
    myList = []

    for line in f:

        myList.append(line)

    return myList
    
myList=read_data(f)
myList
```

```python
def housecount(myList):
    n=0
    for a in myList:
        if 'House ID' in a:
            n=n+1
    return n    
housecount(myList)
```

```python
array=np.zeros((1202,19))

for i in range(0,1202):
    for j in range(0,19):
        array[i][j]=-1
```

```python
def extract(array,myList):
    i=-1
    #d1 = dt.datetime(1000, 1, 1)
    for a in myList:
        if 'House ID' in a:
            i=i+1
            start = a.find(':  ') + 3
            end = a.find('\n', start)
            array[i][0]=int(a[start:end],16)
#        if 'Date Built' in a:
#            start = a.find('ilt : ') + 6
#            end = a.find(' ', start)
#            datobject = datetime.strptime(a[start:end], '%m/%d/%Y')
#            delta = datobject - d1
#            b=delta.days
#            array[i][1]=b
#        if 'Date Priced' in a:
#            start = a.find('ced :  ') + 7
#            end = a.find(' ', start)
#            dat_object = datetime.strptime(a[start:end], '%m/%d/%Y')
#            delta = dat_object - d1
#            b=delta.days
#            array[i][19]=b
        if 'garden' in a:
            if 'beautiful' in a:
                array[i][2]=1
            if 'no space' in a:
                array[i][2]=0
        if 'Dock' in a:
            start = a.find('Dock is ') + 8
            end = a.find(' holy', start)
            array[i][3]=float(a[start:end])
        if 'Capital' in a:
            start = a.find('Capital is ') + 11
            end = a.find(' holy', start)
            array[i][4]=float(a[start:end])
        if 'Market' in a:
            start = a.find('Market is ') + 10
            end = a.find(' holy', start)
            array[i][5]=float(a[start:end])
        if 'Tower' in a:
            start = a.find('Tower is ') + 9
            end = a.find(' holy', start)
            array[i][6]=float(a[start:end])
        if 'River' in a:
            start = a.find('River is ') + 9
            end = a.find(' holy', start)
            array[i][7]=float(a[start:end])
        if 'renovation' in a:
            if 'Mighty King' in a:
                array[i][8]=1
            if 'did not' in a:
                array[i][8]=0
        if 'dining room' in a:
            start = a.find('are ') + 4
            end = a.find(' dini', start)
            array[i][9]=int(a[start:end])
        if 'bedroom' in a:
            start = a.find('are ') + 4
            end = a.find(' bed', start)
            array[i][10]=int(a[start:end])
        if 'bathroom' in a:
            start = a.find('are ') + 4
            end = a.find(' bath', start)
            array[i][11]=int(a[start:end])
        if 'isit' in a:
            if 'great' in a:
                array[i][12]=1
            if 'could' in a:
                array[i][12]=0
        if 'orcerer' in a:
            if 'could' in a:
                array[i][13]=1
            if 'cursed' in a:
                array[i][13]=0
        if 'blessing' in a:
            start = a.find('with ') + 5
            end = a.find('bles', start)
            array[i][14]=int(a[start:end])
        if 'farm' in a:
            if 'huge' in a:
                array[i][15]=2
            if 'small' in a:
                array[i][15]=1
            if 'no' in a:
                array[i][15]=0
        if 'Location' in a:
            if 'Mountains' in a:
                array[i][16]=3
            if 'Cursed' in a:
                array[i][16]=1
            if 'Servant' in a:
                array[i][16]=2
            if 'King' in a:
                array[i][16]=4
        if 'Holy tree' in a:
            if 'tall' in a:
                array[i][17]=1
            if 'death' in a:
                array[i][17]=0
        if 'Knight' in a:
            start = a.find('is ') + 3
            end = a.find(' holy', start)
            array[i][18]=float(a[start:end])
    return array

array=extract(array,myList)
```
